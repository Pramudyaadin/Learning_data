### Data Cleaning

```python
data_new = data_pokemon.dropna()
data_new 
```

```python
data_new.rename(columns= {'#' : 'Number', 'Type 1' : 'Type_1', 'Type 2' : 'Type_2', 'Sp. Atk' : 'Sp_Atk', 'Sp. Def' : 'Sp_Deff'}, inplace= True)
```

```python
data_new
```

### Explorasi Data

#### Query Data

```python
# mengambil data pokemon dengan 2 Type, tetapi Type1 Water
# memiliki Total lebih besar dari 500
# generation lebih dari 3
# data yang ditampilkan hanya number, name, type1, type2, total, dan generasi

data_new.query("Type_1 == 'Water' & Total >= 500 & Generation > 3 ")[[
    'Number', 'Name', 'Type_1', 'Type_2', 'Total', 'Generation']] 
```

```python
# mengambil data pokemon dengan 2 Type, tetapi Type1 Fire
# memiliki attack dan deffense lebih besar dari 80
# tidak termasuk pokemon legendary
# data yang ditampilkan hanya number, name, type1, type2, attack, defense, dan legendary

data_new.query("Type_1 == 'Fire' & Attack >= 80 & Defense >= 80 & Legendary == False ")[[
    'Number', 'Name', 'Type_1', 'Type_2', 'Attack', 'Defense', 'Legendary']] 
```

```python
# mengambil data pokemon yang termasuk katagori legendary
# memiliki attack dan deffense lebih besar dari 100
# memiliki Total lebih dari sama dengan 600
# data yang ditampilkan hanya number, name, type1, type2, Total, dan legendary

data_new.query("Attack >= 100 & Defense >= 100 & Total >= 600 & Legendary == True ")[[
    'Number', 'Name', 'Type_1', 'Type_2', 'Attack', 'Defense', 'Total', 'Legendary']] 
```

#### Grouping Data

```python
# mengetahui pokemon dengan type 1 yang memiliki rataan total lebih dari sama dengan 100

data_new2 = data_new.groupby('Type_1')[['Total']].mean()
data_new2[data_new2['Total'] >= 200]
```

```python
# mengetahui rataan attack dari setiap pokemon pada type 1

data_new.groupby('Type_1')['Attack'].mean()
```

```python
# mengetahui rataan total dari setiap pokemon legendary
# pokemon yang termasuk dalam type 1 fire

data_new[
    (data_new['Type_1']  == 'Fire') &
    (data_new['Legendary'] == True)
].groupby(['Number', 'Name', 'Type_1', 'Legendary'])['Total'].mean()
```
